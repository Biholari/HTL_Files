### Vokabel

* `axis`/`axes` - Dimension
* `dtype` - Datentyp
* `shape` - Tupel, das die Form des Arrays beschreibt

````python
import numpy as np
````

### Erzeugung

````python
array = np.array([42, 3.1415])
print(array, array.shape, array.dtype, type(array))
````

````python
matrix = np.arange(0, 24, 1)
print(matrix)
matrix = matrix.reshape(4, 6)
print(matrix, matrix.shape)
matrix.reshape(3, 2, 2, -1)  # -1 => was sich ausgeht
````

````python
x = np.linspace(0, np.pi, 20)  # für floats besser geeignet als arange
print(x, x.shape)
sin = np.sin(x)
print(sin)
````

### Slicing

* wie bei Listen
* bei mehreren Dimensionen Indices als Tupel
* Index kann selbst Liste sein

````python
matrix = np.arange(0, 12).reshape(3, 4)
matrix = 2 ** matrix
matrix
````

````python
print(matrix[1, 1])
print(matrix[1, :])
print(matrix[1])
print(matrix[:, 1])
print(matrix[::2, 0:2])
print(matrix[[1, 2], 1])
````

### Conditional selection/Boolean indexing

* effektive filter-Methode

````python
matrix = np.arange(0, 12).reshape(3, 4)
print(matrix)
is_even = matrix % 2 == 0
print(is_even)
even_elements = matrix[is_even]
print(even_elements)
````

````python
matrix[matrix > 7]
````
